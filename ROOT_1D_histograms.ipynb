```python
import matplotlib.pyplot as plt
import numpy as np
import ROOT
```

```python
h = ROOT.TH1F("h", "1D histogram", 50, -5.0, 5.0)
h.FillRandom("gaus", 1000)
print(f"Entries: {h.GetEntries()}")
print(f"Mean: {h.GetMean()}")
```

```python
nbins = h.GetNbinsX()
x = np.array([h.GetBinCenter(i) for i in range(1, nbins + 1)])
y = np.array([h.GetBinContent(i) for i in range(1, nbins + 1)])
yerr = np.array([h.GetBinError(i) for i in range(1, nbins + 1)])
plt.errorbar(x, y, yerr=yerr, fmt="o", markersize=3)
plt.xlabel(h.GetXaxis().GetTitle())
plt.ylabel(h.GetYaxis().GetTitle())
plt.title(h.GetTitle())
plt.tight_layout()
plt.show()
```

Determine the maximum bin and its details.

```python
max_bin = h.GetMaximumBin()
max_val = h.GetBinContent(max_bin)
max_err = h.GetBinError(max_bin)
max_centre = h.GetBinCenter(max_bin)
max_low = h.GetXaxis().GetBinLowEdge(max_bin)
max_up = h.GetXaxis().GetBinUpEdge(max_bin)

print(
    "Max bin details:",
    f"bin={max_bin}, content={max_val}, error={max_err}, "
    f"centre={max_centre}, range=[{max_low}, {max_up}]"
)
```
